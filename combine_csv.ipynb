```python
import pandas as pd
```

```python
dose100 = pd.read_csv('Water_100dose.csv')
dose60 = pd.read_csv('Water_60dose.csv')
dose40 = pd.read_csv('Water_40dose.csv')
```

```python
new_col = list(dose40.columns)
new_col
```

```python
i = 1
for col in dose40.columns:
    if col.startswith('Water_'):
        new_col[i] = 'Water_40dose_' + col[6:]
        i += 1
```

```python
new_col
```

```python
i = 1
for c in dose40.columns:
    if c.startswith('Water_'):
        dose40 = dose40.rename(columns = {c: new_col[i]})
        i += 1
```

```python
dose40 = dose40.drop('Unnamed: 0', axis=1)
```

```python
new_col = list(dose60.columns)
i = 1
for col in dose60.columns:
    if col.startswith('Water_'):
        new_col[i] = 'Water_60dose_' + col[6:]
        i += 1
i = 1
for c in dose60.columns:
    if c.startswith('Water_'):
        dose60 = dose60.rename(columns = {c: new_col[i]})
        i += 1
```

```python
dose60 = dose60.drop('Unnamed: 0', axis=1)
```

```python
new_col = list(dose100.columns)
i = 1
for col in dose100.columns:
    if col.startswith('Water_'):
        new_col[i] = 'Water_100dose_' + col[6:]
        i += 1
i = 1
for c in dose100.columns:
    if c.startswith('Water_'):
        dose100 = dose100.rename(columns = {c: new_col[i]})
        i += 1
```

```python
dose100
```

```python
final = pd.concat([dose100, dose60, dose40], axis=1).transpose()
final
```

```python
final.to_csv('Water.csv')
```
